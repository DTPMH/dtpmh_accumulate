Some suggestions to prevent underfitting or overfitting:
1. Add drop out layers.
2. Apply L1/L2 Regularization
3. Use Batch_Normalization.
4. Use ReLu activation function instead of Sigmoid.
5. Early stop. If the epoch is too small, it will be underfitted. while if it's too big it will be overfitted. So choose a proper epoch.
6. Data augmentation, do some rotating, translating, scaling or mirroring to the original data images.
