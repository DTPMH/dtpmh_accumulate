## Exercise 1: Basis Function Q-learning

As seen in the lecture tabular $Q$-learning can intepreted as minimizing the temporal difference (TD) given by

$$
 \frac{1}{2} \text{TD}^2 = \frac{1}{2} \left (Q_\theta(x_t,u_t) - \left (r_t + \gamma \max_u Q_\theta(x_{t+1},u) \right ) \right )^2
$$

This mimization problem can be used to formulate an steepest decent update rule for continous $Q$ functions.  

**a)** Derive the steepest decent update rule which mimizes $\frac{1}{2} \text{TD}^2$ for the parameters $\theta$ where the $Q$-function has the form of a basis function expansion $$ Q_\theta(x_t,u_t) = \phi(x_t) \cdot \theta_{u_t} $$ where $\theta_{u_t}$ is a vector for a given input $u_t$.

A possible basis function for $Q$-learning is the normalized [radial basis function](https://en.wikipedia.org/wiki/Radial_basis_function)  (RBF) which is given by

$$
\psi_i(x_t) = \exp \left ( {-\frac{(x_t - c_i) B (x_t - c_i)^T}{2 \sigma_c^2}} \right )\\
\phi_i(x_t) = \frac{\psi_i(x_t)}{\sum_i \psi_i(x_t)}\\
B = diag(1/\Delta X_1^2, 1/\Delta X_2^2, ...)
$$

where $\Delta X_j$ is the grid spacing in each dimension.

see: https://en.wikipedia.org/wiki/Radial_basis_function_network
